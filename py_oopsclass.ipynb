TASK:

INTENT CATEGORIES:

- Inquiry:
  The sender is seeking information or clarification.
  Example indicators: "Could you tell me...", "I would like to know..."

- Complaint:
  The sender expresses dissatisfaction, frustration, or reports a problem.

- Request:
  The sender asks for a specific action to be taken.

- Follow-up:
  The sender references previous communication and is checking status or progress.

- Thank You:
  The sender expresses gratitude or appreciation.

URGENCY LEVELS:

- High:
  Immediate attention required (mentions deadlines, urgency, time-sensitive issues).

- Medium:
  Action needed within a few days, but not critical.

- Low:
  No clear urgency or informational only.

CLASSIFICATION RULES:

1. Identify one primary_intent (mandatory).
2. Identify secondary_intent only if clearly present; otherwise return null.
3. Use explicit signals in the email — do not infer unstated intent.
4. If multiple intents exist, choose the dominant purpose as primary.
5. Urgency must be inferred from language cues (e.g., "as soon as possible" → High).
6. Provide concise reasoning explaining the decision.
7. Return ONLY valid JSON. No markdown or extra text.

OUTPUT FORMAT:

{
  "primary_intent": "Inquiry/Complaint/Request/Follow-up/Thank You",
  "secondary_intent": "Inquiry/Complaint/Request/Follow-up/Thank You or null",
