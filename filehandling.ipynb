````python
with open("resumes.txt", "w") as file:  # Automatically closes
    file.write("Alice, Python, 5 years\n")
    file.write("Bob, Java, 3 years\n")
# File automatically closed here, even if error occurs!

# Reading with with statement:
with open("resumes.txt", "r") as file:
    content = file.read()          # Read entire file
    lines = file.readlines()       # Read as list of lines
    
# Process line by line (memory efficient):
with open("resumes.txt", "r") as file:
    for line in file:              # Read one line at a time
        print(line.strip())
````

````python
import json

# Python data to save:
candidates = [
    {"name": "Alice", "skills": ["Python"], "exp": 5},
    {"name": "Bob", "skills": ["Java"], "exp": 3}
]

# Save to JSON file:
with open("candidates.json", "w") as file:
    json.dump(candidates, file, indent=2)  # indent=2 for readability

# Load from JSON file:
with open("candidates.json", "r") as file:
    loaded_candidates = json.load(file)    # Back to Python list

print(loaded_candidates[0]["name"])  # "Alice"
````

````python
def save_resume(resume_data, filename="resumes.json"):
    """Save resume data to JSON file"""
    try:
        # Step 1: Try to load existing data
        try:
            with open(filename, 'r') as f:
                resumes = json.load(f)  # Load existing resumes
        except FileNotFoundError:
            resumes = []  # File doesn't exist yet
        
        # Step 2: Add new resume
        resumes.append(resume_data)
        
        # Step 3: Save back to file
        with open(filename, 'w') as f:
            json.dump(resumes, f, indent=2)  # Write with formatting
        
        return True
    except Exception as e:
        print(f"Error: {e}")
        return False
````

Day 9: Industry Pro Tip – File Handling
Always specify encoding when reading/writing text files:

Potential problems with default encoding:

python

Copy code
with open("resume.txt", "r") as f:  # Uses system default encoding
    text = f.read()  # Might fail with non-ASCII characters!
Always specify UTF-8:

python

Copy code
with open("resume.txt", "r", encoding="utf-8") as f:
    text = f.read()  # Safe with all characters
