```python
resumes = [ #inputing the list items with elements
    {"name": "Alice", "skills": ["Python", "SQL", "Django"], "years": 5},
    {"name": "Bob", "skills": ["Java", "Spring", "SQL"], "years": 3},
    {"name": "Carol", "skills": ["Python", "Flask", "React"], "years": 7}
]

required_skills = ["Python", "SQL"]#required elements to be inputed  the analizer for the company team

# Outer loop: Process EACH resume
for resume in resumes:#this is for the outer for loop such a way it checks first with name and diaplays the name first and then goes inside 
    print(f"\n--- {resume['name']} ---")
    
    # Inner loop: Check EACH required skill
    matched_skills = []# we re taking a null list for appending the items before going for the inner loop operation
    for skill in required_skills:# this inner loop checks all the skills of each person with the required skills alloted by the company side 
        if skill in resume["skills"]:
            matched_skills.append(skill)
    
    # Calculate match percentage
    match_percentage = (len(matched_skills) / len(required_skills)) * 100
    print(f"Match: {match_percentage:.0f}%")# formatting cna be done by using{var_name:.0f}% like that
```
