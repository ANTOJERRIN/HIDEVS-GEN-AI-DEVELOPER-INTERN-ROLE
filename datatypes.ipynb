```python
## Day 1 CODEQUEST Example Explained:
name = input("Enter your name: ")       # Gets TEXT → String type
age = int(input("Enter your age: "))    # Converts to NUMBER → Integer type
height = float(input("Enter height: ")) # Converts to DECIMAL → Float type

# type() function shows what "box type" you're using
print(type(name))    # Shows: <class 'str'>  (string box)
print(type(age))     # Shows: <class 'int'>  (integer box)
print(type(height))  # Shows: <class 'float'> (float box)
```
