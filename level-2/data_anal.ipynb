Data Analysis: Finding Stories in Numbers
Data analysis is like being a detective with numbers as clues. Raw data is messy and confusing—like a crime scene. Cleaning organizes the evidence, analysis finds patterns, and insights solve the mystery. Every dataset tells a story; your job is to listen.

The Data Analysis Process
Collection: Gather data from various sources
Cleaning: Fix errors, handle missing values, standardize formats
Exploration: Look for patterns, relationships, anomalies
Analysis: Apply statistical methods, create models
Visualization: Create charts to communicate findings
Interpretation: Translate numbers into business insights
Why Data Analysis Matters for Feedback
Trend Identification: Spot increasing complaints about specific features
Root Cause Analysis: Understand why ratings are dropping
Customer Segmentation: Different feedback from different user groups
Impact Measurement: Quantify effect of product changes on sentiment
Common Analysis Techniques
Descriptive: What happened? (averages, counts)
Diagnostic: Why did it happen? (correlations, comparisons)
Predictive: What will happen? (trends, forecasts)
Prescriptive: What should we do? (recommendations)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def exploratory_analysis(df):
    """Comprehensive EDA for feedback data"""

    print("=== DATA SHAPE ===")
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

    print("\n=== DATA TYPES ===")
    print(df.dtypes)

    print("\n=== MISSING VALUES ===")
    print(df.isnull().sum())

    print("\n=== BASIC STATISTICS ===")
    print(df.describe())

    # Visualizations
    if 'rating' in df.columns:
        plt.figure(figsize=(10, 4))
        plt.subplot(1, 2, 1)
        df['rating'].hist(bins=5)
        plt.title('Rating Distribution')

        plt.subplot(1, 2, 2)
        if 'category' in df.columns:
            df['category'].value_counts().plot(kind='bar')
            plt.title('Feedback Categories')

        plt.tight_layout()
        plt.show()
```

AI Context Windows: The AI's Memory Limit
Context window is like the AI's short-term memory. It can only "remember" a certain amount of text from the conversation. Everything beyond that gets forgotten.

How Context Windows Work
Token-based: AI processes text in tokens (about 4 characters per token)
Fixed Size: Typical context windows: 4K, 8K, 16K, 32K tokens
Sliding Window: New text pushes out old text
Memory Management: Critical for long conversations
Strategies for Managing Context
Summarization: Periodically summarize conversation
Prioritization: Keep important information, discard small talk
External Memory: Store conversation history in database
Chunking: Break long documents into manageable pieces
Real-World Impact
Customer Support: Remember entire conversation history
Document Analysis: Process long reports or contracts
Code Review: Analyze entire codebases
Research: Synthesize information from multiple sources

Answer

Mean is average (sum divided by count), median is middle value when sorted, mode is most frequent value. Use median when data has outliers (like income data). Use mode for categorical data (most common category).
