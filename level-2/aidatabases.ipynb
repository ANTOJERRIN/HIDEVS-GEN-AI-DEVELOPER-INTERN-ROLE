```python
# Using SQLAlchemy ORM
from sqlalchemy import create_engine, Column, Integer, String, DateTime
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
from datetime import datetime

Base = declarative_base()

class Feedback(Base):
    __tablename__ = 'feedback'

    id = Column(Integer, primary_key=True)
    text = Column(String(500), nullable=False)
    rating = Column(Integer, nullable=False)
    category = Column(String(50))
    created_at = Column(DateTime, default=datetime.utcnow)

    def __repr__(self):
        return f"<Feedback(id={self.id}, rating={self.rating})>"

# Setup
engine = create_engine('sqlite:///feedback.db')
Base.metadata.create_all(engine)
Session = sessionmaker(bind=engine)

# Usage
session = Session()
new_feedback = Feedback(text="Great product!", rating=5, category="praise")
session.add(new_feedback)
session.commit()
```

Databases: Organized Storage for Persistent Data
Think of databases as digital filing cabinets. Without databases, data disappears when program closes. With databases, data persists, can be searched efficiently, and maintains relationships between different pieces of information.

Why Use Databases
Persistence: Data survives program restarts and crashes
Efficiency: Fast searching with indexes
Concurrency: Multiple users can access simultaneously
Integrity: Enforce rules (valid ratings, required fields)
Relationships: Connect related data (customers → feedback → responses)
SQLite vs Other Databases
SQLite: File-based, simple, great for learning and small apps
MySQL/PostgreSQL: Server-based, powerful, for production web apps
MongoDB: Document-based, flexible schema, for unstructured data
Core Database Concepts
Tables: Like Excel sheets, hold specific types of data
Rows: Individual records (one feedback item)
Columns: Fields in each record (text, rating, date)
Primary Key: Unique identifier for each row
Foreign Key: Link to row in another table

Few-shot Learning: Teaching AI with Examples
Few-shot learning is like showing someone how to file documents. Instead of explaining rules, you show 2-3 examples of correctly filed documents, then ask them to file new ones. The AI learns the pattern from examples.

Why Few-shot Works
Pattern Recognition: Humans and AI both learn well from examples
Reduced Ambiguity: Shows exactly what you want
Handles Complexity: Examples can show nuanced cases
Flexible: Easy to update by changing examples
Designing Good Examples
Diverse: Show different formats and edge cases
Clear: Examples should be unambiguous
Representative: Typical cases, not rare exceptions
Correct: Must show exactly what you want
When to Use Few-shot
Complex Tasks: Where rules are hard to specify
Formatting: When output format is critical
Edge Cases: To show how to handle unusual inputs
Quality Control: To establish minimum quality standard
