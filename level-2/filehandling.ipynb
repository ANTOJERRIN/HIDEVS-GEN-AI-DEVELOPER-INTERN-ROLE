File Handling: Persistent Data Storage
Files are the most basic form of persistent storage. While databases are more powerful, files are simpler and more universal. Every program needs to read or write files at some point—for configuration, data export, logging, or data import.

File Operations Hierarchy
Open: Get access to the file (like unlocking a drawer)
Read/Write: Work with the content (like reading/writing in a notebook)
Close: Release the file (like locking the drawer)
Error Handling: What if a file doesn't exist or is corrupted?
Common File Formats and Their Uses
Text Files (.txt): Simple data, logs, configuration
CSV Files (.csv): Tabular data, spreadsheets, data exchange
JSON Files (.json): Structured data, APIs, configuration
Binary Files: Images, videos, compiled programs
File Handling Best Practices
Always Use Context Managers: with open() as file: auto-closes
Specify Encoding: Always use encoding='utf-8' for text
Handle Exceptions: Files can be missing, corrupted, or locked
Close Properly: Unclosed files can cause memory leaks

Prompt Refinement: The Art of Getting Better AI Responses
Prompt refinement is an iterative process of improving AI prompts based on results. It's like tuning a radio—small adjustments make the signal clearer. The first prompt is rarely perfect; refinement gets you better results.

The Refinement Process
Initial Prompt: Start with basic request
Test: Run with sample inputs
Analyze Results: What works? What fails?
Refine: Add specificity, examples, constraints
Repeat: Until results meet quality standards
Common Refinement Patterns
Add Specificity: "Extract data" → "Extract name, email, and date"
Provide Examples: Show exactly what you want
Add Constraints: "Must be under 100 words"
Specify Format: "Return as JSON with these fields"
Set Role: "As a senior analyst..."
When to Refine
AI misunderstands the task
Output format is inconsistent
Edge cases aren't handled well
Quality varies too much
New requirements emerge

What's the difference between text mode and binary mode?

Answer

Text mode ('r', 'w') handles encoding/decoding and newline conversions. Binary mode ('rb', 'wb') reads/writes raw bytes. Use text for human-readable files, binary for images, videos, or when you need exact byte control.

Why use `with open()` instead of `open()` and `close()`?
13

sec

Answer

with statement ensures file is closed even if exceptions occur. It's a context manager that calls `__enter__()` on entry and `__exit__()` on exit. Prevents resource leaks and makes code cleaner.

How do you handle large files that don't fit in memory?

Answer

Read line by line: `for line in open('large.txt'): process(line)`. Or read in chunks: `while chunk := f.read(8192): process(chunk)`. Avoid `read()` or `readlines()` for large files.
