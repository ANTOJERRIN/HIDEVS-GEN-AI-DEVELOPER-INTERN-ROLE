```python
import requests
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry

def create_robust_session():
    """Create HTTP session with retry logic"""
    session = requests.Session()

    # Retry strategy
    retry_strategy = Retry(
        total=3,  # Total retries
        backoff_factor=1,  # Wait 1s, 2s, 4s between retries
        status_forcelist=[429, 500, 502, 503, 504],  # Retry on these status codes
    )

    adapter = HTTPAdapter(max_retries=retry_strategy)
    session.mount("http://", adapter)
    session.mount("https://", adapter)

    return session

# Usage
session = create_robust_session()
try:
    response = session.get("https://api.example.com/data", timeout=5)
except requests.exceptions.Timeout:
    print("Request timed out")
```

Understanding Robust HTTP Sessions with Retry Logic
This Python code creates a reliable HTTP session that automatically retries failed requests and handles temporary network issues or server overloads. It's designed to make your API calls more resilient.

**Core Definition:**
The code defines a function, create_robust_session(), which configures a requests library session to automatically retry HTTP requests when certain errors occur, such as server errors or rate limits. It also includes a timeout for individual requests.

**Key Components:**

**requests.Session():**
A Session object allows you to persist certain parameters across requests, like cookies, headers, and, in this case, retry logic. This is more efficient than making individual requests.get() or requests.post() calls.

**Retry Strategy:**
This defines how and when to retry a request.
total=3: Specifies that the request will be retried up to 3 times if it fails.
backoff_factor=1: Introduces a delay between retries. The delay increases exponentially (1 second, then 2 seconds, then 4 seconds).
status_forcelist=[429, 500, 502, 503, 504]: These are the HTTP status codes that will trigger a retry.
429 Too Many Requests: Indicates rate limiting.
5xx codes (e.g., 500 Internal Server Error, 503 Service Unavailable): Indicate server-side issues.

**HTTPAdapter:**
This adapter integrates the retry strategy with the HTTP session. It's like a plug-in that modifies how requests are sent.
session.mount("http://", adapter) / session.mount("https://", adapter):
These lines attach the configured adapter to both http:// and https:// URLs within this session, ensuring the retry logic applies to all requests made through it.

**How It Works:**
1.  A requests.Session object is initialized.
2.  A Retry object is configured to specify retry attempts, delays, and HTTP status codes that should trigger a retry.
3.  An HTTPAdapter is created and linked to this Retry strategy.
4.  The HTTPAdapter is then mounted onto the session for both HTTP and HTTPS protocols.

**5.  When session.get() is called:**
If the request fails with one of the specified status_forcelist codes, the session will automatically wait and retry the request up to total times.
If the request takes longer than the timeout=5 seconds, a requests.exceptions.Timeout error is raised, which is then caught and handled by the except block.

**Why It Matters:**
This pattern significantly improves the robustness of your applications when interacting with external APIs. It reduces the likelihood of your application failing due to transient network glitches, temporary API downtime, or rate limiting, leading to a more stable and user-friendly experience.
