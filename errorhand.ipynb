```python
class FeedbackError(Exception):
    """Base exception for feedback-related errors"""
    pass

class InvalidRatingError(FeedbackError):
    """Raised when rating is out of valid range"""
    pass

class MissingFieldError(FeedbackError):
    """Raised when required field is missing"""
    pass

# Usage
def process_feedback(data):
    if "rating" not in data:
        raise MissingFieldError("rating field is required")
    if not 1 <= data["rating"] <= 5:
        raise InvalidRatingError(f"Rating {data['rating']} must be 1-5")
```
